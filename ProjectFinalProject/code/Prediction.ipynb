```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
prediction = pd.read_csv('toMfinal predict.csv')
prediction['publishedDate'] = pd.to_datetime(prediction['publishedDate'])
```

```python
APPLstock= pd.read_csv("AAPL.csv")
GOOGLstock=pd.read_csv("GOOG.csv")
IBMstock = pd.read_csv("IBM.csv")
AMZNstock = pd.read_csv('AMZN.csv')
MSFTstock = pd.read_csv('MSFT.csv')

APPLstock = APPLstock.drop(APPLstock.index[0:34])
GOOGLstock = GOOGLstock.drop(GOOGLstock.index[0:34])
IBMstock = IBMstock.drop(IBMstock.index[0:34])
MSFTstock = MSFTstock.drop(MSFTstock.index[0:34])
AMZNstock = AMZNstock.drop(AMZNstock.index[0:34])
```

# APPL

```python
dfAPPL = prediction[prediction['APPL_company'] == 1]
dfAPPL = dfAPPL.sort_values(by=['publishedDate'])

appleSenti_list = []
appleScore = list(dfAPPL['APPL_senti'])
for a in range(len(appleScore)):
    score = sum(appleScore[0:a])
    appleSenti_list.append(score)
    
dfAPPL['APPLmovement'] = appleSenti_list

plt.figure(figsize=(20,5),dpi=100)
ax = sns.lineplot(x='publishedDate',y = 'APPLmovement', data =dfAPPL )
plt.title("APPL_Movement")
plt.show()
```

```python
APPLstock['Date'] = pd.to_datetime(APPLstock['Date'])

palette = sns.color_palette("mako_r", 6)
plt.figure(figsize=(20,5),dpi=100)
ax = sns.lineplot(x='Date',y='Close',data = APPLstock)
plt.title('APPL Stock Price Movement')
plt.show()
```
