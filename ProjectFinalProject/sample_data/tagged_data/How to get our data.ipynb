```python
import pandas as pd
```

```python
df1 =  pd.read_csv('kangm.csv',index_col = 0)
df1.head()
df1.shape
```

```python
df2 =  pd.read_csv('alexx.csv',index_col = 0)
df2 = df2.dropna(axis =1)
df2.shape
```

```python
df3 =  pd.read_csv('mxq.csv',index_col = 0)
df3 = df3.dropna(axis =1)
df3.shape
```

```python
d1 = df1.values
d2 = df2.values
d3 = df3.values
```

```python
D = d1+d2 +d3
D.shape
```

```python
for i in range(D.shape[0]):
    for j in range(5):
        if D[i][j] >=2:
            D[i][j] = 1
        else:
            D[i][j] = 0
            
for i in range(D.shape[0]):
    for j in range(5,D.shape[1]):
        if D[i][j] >0:
            D[i][j] = 1
        elif D[i][j] < 0:
            D[i][j] = -1
        else:
            D[i][j] = 0
            
```

```python
colname = list(df1.columns)

dfscore = pd.DataFrame(D,columns=colname)
dfscore
```

```python
df = pd.read_csv('news20190406_285_YeqingXia.csv',index_col = 0)
dfinfo = df.iloc[:,:4]
dfinfo = dfinfo.reset_index().drop('index',axis = 1)
dffianl = pd.concat([dfinfo,dfscore],axis =1 )
```

```python
dffianl.to_csv('finalInput.csv')
```
